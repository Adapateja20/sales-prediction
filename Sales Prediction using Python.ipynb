```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error, r2_score, mean_squared_error

# Load the dataset
try:
    df = pd.read_csv(r"C:\Users\adapa\Downloads\Advertising.csv")
    print("Dataset loaded successfully!")
except FileNotFoundError:
    print("Error: The file 'Advertising.csv' was not found. Please check the path.")

if 'Unnamed: 0' in df.columns:
    df.drop('Unnamed: 0', axis=1, inplace=True)
    print("Dropped 'Unnamed: 0' column.")

plt.figure(figsize=(15, 5))
plt.subplot(1, 4, 1)
sns.histplot(df['TV'], kde=True)
plt.title('TV Ad Budget Distribution')

plt.subplot(1, 4, 2)
sns.histplot(df['Radio'], kde=True)
plt.title('Radio Ad Budget Distribution')

plt.subplot(1, 4, 3)
sns.histplot(df['Newspaper'], kde=True)
plt.title('Newspaper Ad Budget Distribution')

plt.subplot(1, 4, 4)
sns.histplot(df['Sales'], kde=True)
plt.title('Sales Distribution')

plt.tight_layout()
plt.show()

plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
sns.scatterplot(x='TV', y='Sales', data=df)
plt.title('TV Ad Budget vs. Sales')

plt.subplot(1, 3, 2)
sns.scatterplot(x='Radio', y='Sales', data=df)
plt.title('Radio Ad Budget vs. Sales')

plt.subplot(1, 3, 3)
sns.scatterplot(x='Newspaper', y='Sales', data=df)
plt.title('Newspaper Ad Budget vs. Sales')

plt.tight_layout()
plt.show()

sns.pairplot(df, x_vars=['TV', 'Radio', 'Newspaper'], y_vars='Sales', height=4, aspect=1, kind='reg')
plt.suptitle('Advertising Channels vs. Sales (with Regression Line)', y=1.02) 
plt.show()
```
